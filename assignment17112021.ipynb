# Read

```python
df = open("review.txt", "r")
read=df.read()
```

# Tokenization

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize = word_tokenize(read)
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
WordNet_Lemmatizer = WordNetLemmatizer()
lemmatizer = []
```

```python
for word_lemmatize in word_tokenize:
    if word_lemmatize.isalpha():
        lemmatizer.append(WordNet_Lemmatizer.lemmatize(word_lemmatize))
```

```python
print(lemmatizer)
```
