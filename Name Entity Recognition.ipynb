```python
import re
string = data_complete_new
# print(string)
diseases = filtered_sentence
TRAIN_DATA = []
for disease in diseases:
    if len(disease)>3:
        df= re.search(disease, string)
#         print(df)
        while df!=None:
            list_n = [df.start(), df.end(), 'DISEASE']
#             print(list_n)
            TRAIN_DATA.append(tuple(list_n))
            break
# print(TRAIN_DATA)
TRAIN_DATA = [(data_complete_new, {'entities': TRAIN_DATA})]
# print(TRAIN_DATA)
# print(string[50:54])  
# print(string[50:59])  
# print(string[261:266])  
# print(string[160:165])  
# print(string[14249:14253])  
# print(string[437:441])  
# print(string[2873:2879])  

print ((TRAIN_DATA[0][1]['entities']))
```

```python
# Add new Labels to new
for _, annotations in TRAIN_DATA:
  for end in annotations.get('entities'):
    ner.add_label(end[2])
```

```python
# import necessary libraries for training
import random
from spacy.util import minibatch, compounding
from spacy.training import Example
```

```python
pipe_exceptions = ['ner', 'trf_wordpiecer', 'tfg_tok2vec']
unaffected_pipes = [pipe for pipe in nlp.pipe_names if pipe not in pipe_exceptions]
```

```python
with nlp.disable_pipes(*unaffected_pipes):
  sizes = compounding(1.0, 4.0, 1.001)
  
  for itn in range(100):
    random.shuffle(TRAIN_DATA)
    batches = minibatch(TRAIN_DATA, size=sizes)

    losses = {}
    for batch in batches:
      for text, annotations in batch:
        doc = nlp.make_doc(text)
        example = Example.from_dict(doc, annotations)
        nlp.update([example], drop=0.5, losses=losses)
        print("losses: ", losses)
```
